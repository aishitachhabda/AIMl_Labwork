```python
# CELL 1
# heapq provides a PRIORITY QUEUE using a "min-heap"
# In A*, OPEN is a priority queue where the node with smallest f is selected first.

import heapq
```

```python
# CELL 2
# Graph representation (Weighted Adjacency List)
# Each node stores: [(neighbor1, cost1), (neighbor2, cost2), ...]
# 'cost' means the real path cost to move from current node to neighbor.

graph = {
    'A': [('B', 11), ('C', 14), ('D', 7)],
    'B': [('A', 11), ('E', 15)],
    'C': [('A', 14), ('E', 8), ('F', 10)],
    'D': [('A', 7), ('F', 25)],
    'E': [('B', 15), ('C', 8), ('H', 9)],
    'F': [('C', 10), ('D', 25), ('G', 20)],
    'H': [('E', 9), ('G', 10)],
    'G': [('F', 20), ('H', 10)]
}

# Define start and goal
start = 'A'
goal = 'G'

# Quick sanity prints (to ensure graph is loaded correctly)
print("Start:", start, "| Goal:", goal)
print("Neighbors of A:", graph['A'])   # Should show B(11), C(14), D(7)
print("Neighbors of G:", graph['G'])   # Should show F(20), H(10)
```

```python
# CELL 3
# Heuristic h(n): estimated cost from node n to the goal (G).
# These are given in your diagram (red numbers).
# Important: h(goal) MUST be 0.

h = {'A': 40, 'B': 32, 'C': 25, 'D': 35, 'E': 19, 'F': 17, 'H': 10, 'G': 0}

# Quick sanity check
print("Heuristic dictionary h(n):", h)
print("Check: h(A) =", h['A'], "| h(G) =", h['G'], "(goal should be 0)")
```

```python
# CELL 4
# Reconstruct the path using parent pointers:
# parent[child] = parent_node
#
# Example:
# parent['G']='H', parent['H']='E', parent['E']='C', parent['C']='A', parent['A']=None
# Then path is: A -> C -> E -> H -> G

def reconstruct_path(parent, start, goal):
    # If goal was never reached/discovered, we cannot reconstruct a path
    if goal not in parent:
        return None

    path = []            # will store nodes on the path
    current = goal       # begin from goal and go backward to start

    # Keep going backward until we reach None (parent of start)
    while current is not None:
        path.append(current)        # add current node
        current = parent[current]   # move to its parent

    # Currently path is in reverse order (goal -> ... -> start), so reverse it
    path.reverse()

    return path
```

```python
# CELL 5
# A* Search Algorithm
#
# Uses:
# OPEN   = priority queue (min-heap) of (f, g, node)
# visited = CLOSED set (expanded nodes)
# parent = to reconstruct path
# g_cost = best known path cost from start to node (g value)
#
# A* selects the node with smallest f(n) = g(n) + h(n)

def astar(graph, start, goal, h):
    # OPEN is stored as a heap list; heapq keeps the smallest f at the top
    OPEN = []

    # visited (CLOSED) stores nodes we have already expanded fully
    visited = set()

    # parent dictionary for reconstruction; start has no parent
    parent = {start: None}

    # g_cost dictionary stores best known g(n) from start to each node
    g_cost = {start: 0}

    # Push start into OPEN
    # For start: g(start)=0, f(start)=g+h = 0 + h[start] = h[start]
    heapq.heappush(OPEN, (h[start], 0, start))

    # Loop until OPEN is empty (means no nodes left to explore)
    while OPEN:
        # Pop the node with the smallest f
        f_current, g_current, current = heapq.heappop(OPEN)

        # If current was already expanded earlier, skip it (outdated heap entry)
        # This happens because we may push the same node more than once with better costs
        if current in visited:
            continue

        # Minimal print: show which node is selected next by A*
        print("Selected:", current, "| g =", g_current, "| f =", f_current)

        # Goal test: if we popped the goal, we stop
        if current == goal:
            break

        # Mark current as expanded (CLOSED)
        visited.add(current)

        # Expand neighbors of current node
        for neighbor, cost in graph[current]:

            # If neighbor already expanded, ignore it
            if neighbor in visited:
                continue

            # Compute new g value if we go to neighbor via current
            # new_g = g(current) + edge_cost(current, neighbor)
            new_g = g_cost[current] + cost

            # Update rule:
            # If neighbor not seen before OR we found a cheaper path to neighbor
            if (neighbor not in g_cost) or (new_g < g_cost[neighbor]):
                # Update best g for neighbor
                g_cost[neighbor] = new_g

                # Update parent pointer: best way to reach neighbor is from current
                parent[neighbor] = current

                # Compute new f = g + h
                new_f = new_g + h[neighbor]

                # Push neighbor into OPEN with its (f, g, node)
                heapq.heappush(OPEN, (new_f, new_g, neighbor))

    # After loop ends, reconstruct path using parent pointers
    path = reconstruct_path(parent, start, goal)

    # Return path + cost + maps (useful for teaching/debugging)
    return path, g_cost.get(goal, None), parent, g_cost
```

```python
# CELL 6
# Run A* and display final outputs

path, total_cost, parent, g_cost = astar(graph, start, goal, h)

print("\n--- FINAL OUTPUT ---")
print("Path (start -> goal):", path)
print("Total cost of path:", total_cost)

print("\nParent map (used for reconstruction):")
print(parent)

print("\ng_cost map (best known costs from start):")
print(g_cost)
```
